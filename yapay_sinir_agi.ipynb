```python
# Bu kod bir yapay sinir ağını tanımlar.
#Sabitler : w1, w2, w3, w4, w5, w6: Ağırlık katsayılarıdır. Girişlerin her birine ve ara katmanlara uygulanan ağırlıklardır.
#Sabitler : b1, b2, b3: Her bir nöron için kullanılan sabitler (bias)dir. Bu, ağın çıktısını kaydırmak için kullanılır.
w1 = 1
w2 = 0.5
w3 = 1
w4 = -0.5
w5 = 1
w6 = 1

b1 = 0.5
b2 = 0
b3 = 0.5

def l1f1(x1, x2): # İlk katmanda, x1 ve x2 girişleriyle ve w1, w3 ağırlıkları ile b1 sabiti kullanılarak hesaplanan bir nöronun çıkışı.
  return x1 * w1 + x2 * w3 + b1

def l1f2(x1, x2): #l1f2(x1, x2): İlk katmanda, x1 ve x2 girişleriyle ve w2, w4 ağırlıkları ile b2 sabiti kullanılarak hesaplanan ikinci nöronun çıkışı.
  return x1 * w2 + x2 * w4 + b2

def of1(x1, x2): #of1(x1, x2): Çıkış katmanında, l1f1 ve l1f2 nöronlarından gelen değerlerle ve w5, w6 ağırlıkları ile b3 sabiti kullanılarak hesaplanan nöronun çıkışı.
  return x1 * w5 + x2 * w6 + b3

i1 = 1 #1. giriş değeri
i2 = 0 #2. giriş değeri

mustafa_hoca = l1f1(i1, i2) #İlk katmandaki l1f1 fonksiyonunun çıkışı, yani l1f1(1, 0).
ali_hoca = l1f2(i1, i2) #İlk katmandaki l1f2 fonksiyonunun çıkışı, yani l1f2(1, 0).

final_deger = of1(mustafa_hoca, ali_hoca) # Çıkış katmanındaki of1 fonksiyonunun sonucu, yani of1(mustafa_hoca, ali_hoca).

print(final_deger) #Ekrana yazdırılır.
```

```python
import numpy as np

def sigmoid(x): #Sigmoid fonksiyonu, genellikle yapay sinir ağlarında aktifleşme fonksiyonu olarak kullanılır.
  return 1 / (1 + np.exp(-x))

print(sigmoid(-14)) #-14 değeri için sigmoid sonucu ekrana yazdırılır.
```

```python
# Sisteme yeni bir hoca ekleme
#w1, w2, w3, w4, w5, w6: Sinir ağındaki bağlantıların ağırlıklarıdır. Bu ağırlıklar, girişlerin her bir katmandaki nöronlara etkisini belirler.
w1 = 0.7
w2 = 0.3
w3 = 0.4
w4 = 0.6
w5 = 0.55
w6 = 0.45 

#İlk Katman: l1f1 ve l1f2 Fonksiyonları
#l1f1 ve l1f2 fonksiyonları, giriş değerlerini (x1 ve x2) alır ve ağırlıklarla çarpar. İlk katmanda iki farklı nöronun hesaplamalarını yapar.
#l1_sonuc, her bir nöronun lineer kombinasyonudur (x1 * w1 + x2 * w3 veya x1 * w2 + x2 * w4).
#Sonuçlar sigmoid() fonksiyonu ile aktivasyona sokulur ve 0 ile 1 arasında bir değere dönüştürülür.

def l1f1(x1, x2):
  l1_sonuc = x1 * w1 + x2 * w3
  return sigmoid(l1_sonuc)

def l1f2(x1, x2):
  l1_sonuc = x1 * w2 + x2 * w4
  return sigmoid(l1_sonuc)

#Çıkış Katmanı: of1 Fonksiyonu
#of1 fonksiyonu, ilk katmandan gelen iki aktivasyonu (mustafa_hoca ve ali_hoca) alır ve çıkış katmanında bu değerlerin ağırlıklarla çarpılıp toplandığı lineer kombinasyonu hesaplar.
#Sonuç, sigmoid() fonksiyonu ile aktif hale getirilir.

def of1(x1, x2):
  of_sonuc = x1 * w5 + x2 * w6
  return sigmoid(of_sonuc)

#Giriş Verileri ve Hesaplamalar

i1 = 0.5
i2 = 0.3

mustafa_hoca = l1f1(i1, i2)
ali_hoca = l1f2(i1, i2)

final_deger = of1(mustafa_hoca, ali_hoca)

print(round(final_deger, 3)) #Sonuç, round(final_deger, 3) ile üç ondalık basamağa yuvarlanarak ekrana basılır.
```

```python
# Sisteme yeni bir hoca ekleme

#Bu kod, sinir ağı modelinde giriş, ağırlık matrisleri ve aktivasyon fonksiyonları kullanılarak nihai bir çıkış hesaplamaktadır.

matrix_i = np.array([[0.5, 0.3]]) #Matrisin tanımlanması.Burada, iki giriş değeri (0.5 ve 0.3) içeren 1x2'lik bir matris oluşturuluyor.

matrix_l1w = np.array([[0.7, 0.3, 0.5],
                       [0.4, 0.6, 0.5]]) #İlk katmanın ağırlık matrisidir. Bu matris 2x3'lük olup, girişten ara katmanlara giden bağlantıların ağırlıklarını içerir. Burada 2 giriş ve 3 nöron için ağırlıklar tanımlanmıştır.

matrix_ow = np.array([[0.55],
                      [0.45],
                      [0.5]]) #Çıkış katmanının ağırlık matrisidir. Bu matris 3x1'lik olup, ara katmandan çıkışa giden bağlantıların ağırlıklarını içerir.

#Ağırlık toplamlarının hesaplanması.
o1 = np.dot(matrix_i, matrix_l1w)
print(o1)
#matrix_i (giriş matrisini) ve matrix_l1w (ilk katmanın ağırlık matrisini) çarparak ilk katmandaki nöronların ağırlıklı toplamını (o1) hesaplar. np.dot() fonksiyonu matris çarpımını gerçekleştirir.
#Sonuç olarak, o1 değişkeni 1x3'lük bir matris olacaktır.

#Sigmoid Aktivasyonu
so1 = sigmoid(o1) #o1 matrisine sigmoid aktivasyon fonksiyonu uygulanarak, ilk katmandaki nöronların çıktıları so1 olarak hesaplanır.

#Çıkış Katmanının Hesaplanması
o = np.dot(so1, matrix_ow) #so1 (ilk katmandan çıkan değerler) ve matrix_ow (çıkış katmanının ağırlık matrisi) çarparak çıkış katmanındaki nöronların ağırlıklı toplamını hesaplar.

#Çıkış Katmanına Sigmoid Uygulaması
so = sigmoid(o) #Çıkış katmanındaki ağırlıklı toplam (o) üzerine sigmoid fonksiyonu uygulanarak nihai çıkış so hesaplanır.

#Sonucun Yuvarlanması ve Yazdırılması
round(so[0][0], 3) #so matrisindeki çıkış değeri (so[0][0]) üç ondalık basamağa yuvarlanır ve ekrana yazdırılır.
```

```python
import numpy as np

#Sigmoid Fonksiyonu ve Türevi : Sigmoid fonksiyonunun türevi feri yayılım için kullanılır. 

def sigmoid(x):
    # Sigmoid fonksiyonu: 1 / (1 + e^(-x))
    # türevi: -(1 + e^(-x - 1))^2
    return 1 / (1 + np.exp(-x))

def sigmoid_derivative(x):
    
    # Sigmoid'in türevi: s(x) * (1 - s(x))
    return sigmoid(x) * (1 - sigmoid(x))


# İleri yayılım fonksiyonları
# l1f1 ve l1f2: İlk gizli katmanın nöronları için ağırlıklı toplamları hesaplar.
# of1: Çıkış katmanının ağırlıklı toplamını hesaplar.

def l1f1(x1, x2):
    return x1 * w1 + x2 * w3

def l1f2(x1, x2):
    return x1 * w2 + x2 * w4

def of1(x1, x2):
    return x1 * w5 + x2 * w6

# Girişler ve hedef değer
i1 = 0.5 #1. giriş değeri
i2 = 0.3 #2. giriş değeri
y = 1  # Hedef değer

# İleri yayılım
#Giriş verileri üzerinden önce gizli katman, sonra çıkış katmanına kadar ileri yayılım yapılır.

h1_input = l1f1(i1, i2)
h1_output = sigmoid(h1_input)

h2_input = l1f2(i1, i2)
h2_output = sigmoid(h2_input)

h3_input = of1(h1_output, h2_output)
h3_output = sigmoid(h3_input)

# Kayıp fonksiyonu (MSE)
#Çıkış ile hedef değer arasındaki farkın karesini hesaplar.

loss = (h3_output - y) ** 2
print("Çıkış:", h3_output)
print("Kayıp:", loss)

# Geri yayılım
# Çıkış katmanı türevleri

dL_dh3 = 2 * (h3_output - y)  # Kayıp fonksiyonunun h3'e göre türevi
print("dL_dh3:", dL_dh3)
dh3_input = sigmoid_derivative(h3_input)  # Sigmoid türevi
dL_dh3_input = dL_dh3 * dh3_input  # Zincir kuralı

#dL_dh3: Kayıp fonksiyonunun çıkış katmanının çıktısına göre türevidir.
#dh3_input: Çıkış katmanının aktivasyon fonksiyonunun türevidir.
#dL_dh3_input: Çıkış katmanının girdisine göre kayıp fonksiyonunun türevidir.

# Çıkış katmanı ağırlıklarının türevleri
#Çıkış katmanındaki ağırlıkların türevleri hesaplanır.
dL_dw5 = dL_dh3_input * h1_output
dL_dw6 = dL_dh3_input * h2_output

# Çıkış katmanı gizli katman çıkışlarına etkisi
dL_dh1_output = dL_dh3_input * w5
dL_dh2_output = dL_dh3_input * w6

# Gizli katman türevleri
dh1_input = sigmoid_derivative(h1_input)
dh2_input = sigmoid_derivative(h2_input)

dL_dh1_input = dL_dh1_output * dh1_input
dL_dh2_input = dL_dh2_output * dh2_input

# Gizli katman ağırlıklarının türevleri
dL_dw1 = dL_dh1_input * i1
dL_dw3 = dL_dh1_input * i2

dL_dw2 = dL_dh2_input * i1
dL_dw4 = dL_dh2_input * i2

print("\nAğırlık türevleri:")
print("dL_dw1:", dL_dw1)
print("dL_dw2:", dL_dw2)
print("dL_dw3:", dL_dw3)
print("dL_dw4:", dL_dw4)
print("dL_dw5:", dL_dw5)
print("dL_dw6:", dL_dw6)

# Ağırlıkların güncellenmesi
#Ağırlıklar, öğrenme oranı (learning_rate) kullanılarak güncellenir.

learning_rate = 0.1
w1 -= learning_rate * dL_dw1
w2 -= learning_rate * dL_dw2
w3 -= learning_rate * dL_dw3
w4 -= learning_rate * dL_dw4
w5 -= learning_rate * dL_dw5
w6 -= learning_rate * dL_dw6

#Çıktı
#Güncellenmiş ağırlıklar ekrana yazdırılır.
print("\nGüncellenmiş Ağırlıklar:")
print("w1:", w1)
print("w2:", w2)
print("w3:", w3)
print("w4:", w4)
print("w5:", w5)
print("w6:", w6)
```
